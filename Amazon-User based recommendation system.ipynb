```python
from sklearn.metrics import pairwise_distances
import numpy as np
```

```python
movie_similarity = pairwise_distances( df, metric="cosine" )
np.fill_diagonal( movie_similarity, 0 ) 


ratings_matrix = pd.DataFrame(movie_similarity)

ratings_matrix.head(20)
```
